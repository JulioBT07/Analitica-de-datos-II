##*****FILTROS PARA VARIABLES CUALITATIVAS*****

##FILTRO 1

```python
#Filtro por objeto en donde filtramos aquellos productos en donde la columna nombre sea igual a LETICIA RAMIREZ HERNANDEZ
filtro1 = df[df["NOMBRE"] == "LETICIA RAMIREZ HERNANDEZ"]
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Filtro1.cvs")
```

##FILTRO 2

```python
#Filtro por filas donde escogemos la fila 8, 40 y 1200
filtro2=df.iloc[ [8, 40, 1200]  , : ] #Filas no consecutivas
filtro2
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv("Filtro2.cvs")
```

##FILTRO 3

```python
#Filtro por columnas en donde escogemos la columna 2, 5 y 7
filtro3 = df.iloc[ : , [2,5,7]]
filtro3
```

```python
#Convertir archivo filtrado a CSV
filtro3.to_csv("Filtro3.cvs")
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

##Filtro 6

```python
df.info()
```

```python
#Convertimos valores con tipo objeto a valores numericos
df[['PRECIO_UNITARIO',	'COSTO_UNITARIO',	'COSTO_UNITARIO_CALCULADO',	'SUBTOTAL_PARTIDA',	'COSTO_TOTAL_CALCULADO', 'MARGEN_UNITARIO_CALCULADO',	'MARGEN_TOTAL_CALCULADO']] = df[['PRECIO_UNITARIO',	'COSTO_UNITARIO',	'COSTO_UNITARIO_CALCULADO',	'SUBTOTAL_PARTIDA',	'COSTO_TOTAL_CALCULADO', 'MARGEN_UNITARIO_CALCULADO',	'MARGEN_TOTAL_CALCULADO']].replace('[\$,]', '', regex=True).astype(float)
df.info()
```

```python
#Filtro por comparacion para aquellos productos en donde el margen total calculado sea mayor a 1000
filtro6 = df[df["MARGEN_TOTAL_CALCULADO"] >  1000]
filtro6
```

```python
#Convertir archivo filtrado a CSV
filtro6.to_csv("Filtro6.cvs")
```

##FILTRO 7

**FILTROS CON CONECTORES LOGICOS**

```python
#Filtro Y para obtener los produtos con margen total calculado menor a 800 y que el nombre sea igual a Diego Armando Leon Duran
filtro7 = df[(df["MARGEN_TOTAL_CALCULADO"] < 800) & (df["NOMBRE"] == "DIEGO ARMANDO LEON DURAN")]
filtro7
```

```python
#Convertir archivo filtrado a CSV
filtro7.to_csv("Filtro7.cvs")
```

##FILTRO 8

```python
#Filtro o para obtener los productos con costo unitario calculado mayor a 200 o que el nombre sea igual a AARON NUÑO CORDOVA
filtro8 = df[(df["COSTO_UNITARIO_CALCULADO"] > 200)  |  (df["NOMBRE"] == "AARON NUÑO CORDOVA")]
filtro8
```

```python
#Convertir archivo filtrado a CSV
filtro8.to_csv("Filtro8.cvs")
```

##FILTRO 9

```python
#Filtro not para discriminar aquellos productos con PRECIO UNITARIO mayor a 6000 y Cantidad igual a 3
filtro9 = df[~(df["PRECIO_UNITARIO"] >= 6000)  &  ~(df["CANT"] == 3)]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("Filtro9.cvs")
```
